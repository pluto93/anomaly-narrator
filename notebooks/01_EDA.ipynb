```python
# 1. Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For better plot display in notebooks
%matplotlib inline
sns.set(style="whitegrid")

# 2. Load a Sample of the Data (optional: limit rows for performance)
df = pd.read_csv('synthetic_fraud_data.csv', nrows=100000)
print("Loaded shape:", df.shape)

# 3. Basic Preview
display(df.head(5))
display(df.describe(include='all').transpose())

# 4. Check for nulls
df.info()
print(df.isnull().sum().sort_values(ascending=False))
```

```python
# Fraudulent vs legitimate transaction count
sns.countplot(data=df, x='is_fraud')
plt.title("Fraudulent vs Non-Fraudulent Transactions")
plt.xlabel("Is Fraud")
plt.ylabel("Count")
plt.show()

fraud_ratio = df['is_fraud'].mean()
print(f"Fraudulent transactions: {fraud_ratio:.2%}")
```

```python
# Count of fraudulent vs legitimate
sns.countplot(data=df, x='is_fraud')
plt.title("Fraudulent vs Non-Fraudulent Transactions")
plt.show()

fraud_ratio = df['is_fraud'].mean()
print(f"Fraudulent transactions: {fraud_ratio:.2%}")
```

```python
# Parse timestamp
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df['hour'] = df['timestamp'].dt.hour
df['dayofweek'] = df['timestamp'].dt.dayofweek

# Transaction volume by hour
sns.histplot(data=df, x='hour', bins=24, hue='is_fraud', multiple='stack')
plt.title("Transactions by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Count")
plt.show()

# Weekend vs weekday fraud
sns.countplot(data=df, x='weekend_transaction', hue='is_fraud')
plt.title("Weekend vs Weekday Fraud")
plt.xlabel("Weekend Transaction")
plt.ylabel("Count")
plt.show()
```

```python
# Distribution of transaction amounts
sns.histplot(df['amount'], bins=100, kde=True)
plt.title("Transaction Amount Distribution")
plt.xlim(0, df['amount'].quantile(0.99))  # Trim outliers
plt.show()

# Fraud amount comparison
sns.boxplot(x='is_fraud', y='amount', data=df)
plt.yscale('log')
plt.title("Transaction Amount by Fraud Label")
plt.show()
```

```python
# Fraud rate by country
fraud_by_country = df.groupby('country')['is_fraud'].mean().sort_values(ascending=False)
fraud_by_country.head(10).plot(kind='bar', title='Fraud Rate by Country')
plt.ylabel("Fraud Rate")
plt.show()

# Fraud by channel
sns.countplot(data=df, x='channel', hue='is_fraud')
plt.title("Transaction Channel by Fraud Status")
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation matrix for numeric features
numeric_cols = ['amount', 'transaction_hour', 'is_fraud']
corr_df = df[numeric_cols].dropna()

plt.figure(figsize=(6, 4))
sns.heatmap(corr_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
